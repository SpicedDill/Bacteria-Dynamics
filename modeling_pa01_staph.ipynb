# Logistic Growth Model for a P. Aeruginosa 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the logistic growth model
def logistic_growth(t, N0, Nmax, r):
    """
    Logistic growth model.
    t: Time (days)
    N0: Initial cell concentration (CFU/mg)
    Nmax: Maximum cell concentration (CFU/mg)
    r: Growth rate (day^-1)
    """
    return Nmax / (1 + ((Nmax - N0) / N0) * np.exp(-r * t))


N0 = 1e3        # Initial concentration (CFU/mg)
Nmax = 1.67e13  # Maximum concentration (CFU/mg)
r = np.log(2) / 0.45  # Growth rate days (approx 25 - 35 min)
total_mg = 0.8


time = np.linspace(0, 30, 1000)  
CFU = logistic_growth(time, N0, Nmax, r)


threshold = 0.999 * Nmax 
threshold_index = np.where(CFU >= threshold)[0][0]
threshold_time = time[threshold_index]
threshold_cfu = CFU[threshold_index]


plt.figure(figsize=(8, 6))
plt.plot(time, CFU, color='blue', label='Theoretical Concentration')
plt.axvline(x=threshold_time, color='red', linestyle='--', label='Carrying Capacity Threshold')
plt.annotate(f"Carrying Capacity\nReached at {threshold_time:.2f} days",
             xy=(threshold_time, threshold_cfu),
             xytext=(threshold_time - 5, threshold_cfu / 2),
             arrowprops=dict(facecolor='black', arrowstyle="->"),
             fontsize=10)


plt.xlabel('Time (days)')
plt.ylabel('Cell Concentration (CFU/mg)')
plt.title('Logistic Growth Model for Pseudomonas aeruginosa')
plt.legend()
plt.grid()
plt.show()

print(f"Carrying capacity (99.9% of Nmax) reached at {threshold_time:.2f} days.")
print(f"Cell concentration at carrying capacity: {threshold_cfu:.2e} CFU/mg.")


time_early = np.linspace(0, 1, 1000)
CFU = logistic_growth(time_early, N0, Nmax, r)

plt.figure(figsize=(8, 6))
plt.plot(time_early, CFU, color='red', label='Theoretical Concentration')
plt.xlabel('Time (days)')
plt.ylabel('Cell Concentration (CFU/mg)')
plt.title('Logistic Growth Model for Pseudomonas aeruginosa')
plt.legend()
plt.grid()
plt.show()

print("Parameters:")
print(f"Initial concentration (N0): {N0:.2e} CFU/mg")
print(f"Maximum concentration (Nmax): {Nmax:.2e} CFU/mg")
print(f"Growth rate (r): {r:.2f} hour^-1")

plt.show()

"""
Costerton, J. W., Stewart, P. S., & Greenberg, E. P. (1999). Bacterial biofilms: A common cause of persistent infections. Science, 284(5418), 1318–1322. https://doi.org/10.1126/science.284.5418.1318

Jørgensen, K. M., Wassermann, T., Johansen, H. K., & Molin, S. (2015). Diverse adaptive mechanisms in populations of Pseudomonas aeruginosa infecting airways of cystic fibrosis patients. Frontiers in Microbiology, 6, 971. https://doi.org/10.3389/fmicb.2015.00971
"""

```

# Carrying Capacity Calculation

## Calculating the Carrying Capacity $N_{\text{max}}$ for Bacteria

To calculate the carrying capacity $N_{\text{max}}$ for bacteria based on the amount of agar (in mg) in a Petri dish, we can use the following logic:

### Steps

1. **Assume a Bacterial Biomass Yield $(Y)$**:  

   This is the amount of bacterial biomass (e.g., in mg) produced per unit of substrate consumed (e.g., mg agar).  
   - Typical yields for bacteria range from **0.2 to 0.6 mg of biomass per mg of substrate**.

2. **Estimate the Biomass Capacity**:  
   $$
   \text{Maximum Biomass (mg)} = Y \times \text{Total Substrate (mg)}
   $$

3. **Convert Biomass to CFU**:  
   Use an approximate conversion factor based on bacterial cell size and dry weight.  
   - **Dry weight per cell**: ~$10^{-12}$ g = $10^{-9}$ mg (varies slightly by species).  
   $$
   N_{\text{max}} = \frac{\text{Maximum Biomass (mg)}}{\text{Dry weight per cell (mg)}}
   $$

---

### Known Parameters

- **Agar amount in a Petri dish $(S)$**: e.g., $(S)$ mg -- Around $ 0.8$ mg.  
- **Biomass yield $(Y)$**: Estimated from literature or experiments.  
- **Dry weight per cell $(w)$**: Varies by bacterial species.

---

### Formula
$$
N_{\text{max}} = \frac{Y \cdot S}{w}
$$




## Adjustments for Specific Bacteria

### Yield Factor $(Y)$:
- **Pseudomonas aeruginosa**: ~2.085 mg biomass/mg agar (aerobic conditions).
- **Staphylococcus aureus**: ~2.085 mg biomass/mg agar.

Kim, J., Park, W., & Kim, S. (2012). Estimation of biomass concentration and dry cell weight using optical density in Pseudomonas aeruginosa. African Journal of Microbiology Research, 6(4), 491–497. https://doi.org/10.5897/AJMR11.1398

### Dry Weight per Cell $(w)$:
- **Pseudomonas aeruginosa**: ~$1.0 - 2.8 \times 10^{-10}$ mg/cell.

Kim, J., Park, W., & Kim, S. (2012). Estimation of biomass concentration and dry cell weight using optical density in Pseudomonas aeruginosa. African Journal of Microbiology Research, 6(4), 491–497. https://doi.org/10.5897/AJMR11.1398

- **Staphylococcus aureus**: ~$2.8 \times 10^{-10}$ mg/cell.

Wyatt, P. J. (1970). Cell wall thickness, size distribution, refractive index ratio and dry weight content of living bacteria (Staphylococcus aureus). Nature, 226(5242), 277–279. https://doi.org/10.1038/226277a0


```python
def calculate_carrying_capacity(agar_mg, yield_factor, dry_weight_per_cell):
    """
    Calculate the carrying capacity (N_max) based on agar amount, yield, and cell dry weight.

    Parameters:
    - agar_mg (float): Amount of agar in mg.
    - yield_factor (float): Biomass yield (mg biomass per mg agar).
    - dry_weight_per_cell (float): Dry weight per bacterial cell in mg.

    Returns:
    - N_max (float): Estimated carrying capacity (CFU).
    """
    max_biomass = yield_factor * agar_mg  # Total biomass supported by the agar
    N_max = max_biomass / dry_weight_per_cell  # Convert biomass to CFU
    return N_max


agar_mg = 800  
yield_factor = 2.085  
dry_weight_per_cell_pa01 = 1e-10  # ~1e-12 g = 1e-9 mg per bacterial cell
dry_weight_per_cell_staph = 2.8e-10

N_max_pseudomonas = calculate_carrying_capacity(agar_mg, yield_factor, dry_weight_per_cell_pa01)
print(f"Estimated carrying capacity for Pseudomonas aeruginosa: {N_max_pseudomonas:.2e} CFU")

N_max_staph = calculate_carrying_capacity(agar_mg, yield_factor, dry_weight_per_cell_staph)
print(f"Estimated carrying capacity for Staphylococcus aureus: {N_max_staph:.2e} CFU")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def bacterial_decay_reversed(dose, max_cf, half_max_dose, steepness):
    """
    Logistic function for bacterial decay with reversed trend.
    """
    return max_cf / (1 + np.exp(steepness * (dose - half_max_dose)))

def error_function(params, target_dose, target_reduction, max_cf):
    """
    Calculate the error between the logistic curve and the target point.
    """
    half_max_dose, steepness = params
    predicted_cfu = bacterial_decay_reversed(target_dose, max_cf, half_max_dose, steepness)
    predicted_reduction = 100 * (1 - predicted_cfu / max_cf)
    return (predicted_reduction - target_reduction) ** 2


max_cf = 100  # Initial CFU count
initial_half_max_dose = 2.2  # Initial guess for half-max dose
initial_steepness = 8.0  # Initial guess for steepness

target_dose_1, target_reduction_1 = 2.35, 95.42  
target_dose_2, target_reduction_2 = 2.2, 86.41  

result = minimize(
    error_function,
    x0=[initial_half_max_dose, initial_steepness],
    args=(target_dose_1, target_reduction_1, max_cf),
    bounds=[(0, None), (0, None)] 
)

optimized_half_max_dose, optimized_steepness = result.x

doses = np.linspace(0, 5, 100)
cfu_counts = bacterial_decay_reversed(doses, max_cf, optimized_half_max_dose, optimized_steepness)
reductions = 100 * (1 - cfu_counts / max_cf)

plt.figure(figsize=(8, 5))
plt.plot(doses, reductions, label="CFU Reduction Curve", color="blue")
plt.axvline(target_dose_1, color="green", linestyle="--", label="Target Dose 1 (2.35 mCi)")
plt.axvline(target_dose_2, color="purple", linestyle="--", label="Target Dose 2 (2.2 mCi)")
plt.scatter([target_dose_1], [target_reduction_1], color="red", label="Target Point 1 (95.42%)")
plt.scatter([target_dose_2], [target_reduction_2], color="orange", label="Target Point 2 (86.41%)")
plt.xlabel("Radiation Dose (mCi)")
plt.ylabel("CFU Reduction (%)")
plt.title("Bacteria Reduction vs. Radiation Dose")
plt.legend()
plt.grid(True)
plt.show()


print(f"Optimized Half-Max Dose: {optimized_half_max_dose:.4f}")
print(f"Optimized Steepness: {optimized_steepness:.4f}")


```

# General Model of Sulphur-35 Decay via Beta Emission

```python
import numpy as np
import matplotlib.pyplot as plt


n0 = 2.5  # Initial dose in mCi
t_half = 87.4  # Half-life of S-35 in days
time = np.linspace(0, 500, 500)  # Time range in days

# decay formula
n_t = n0 * (0.5) ** (time / t_half)

day1, day2 = 8, 16
rad1 = n0 * (0.5) ** (day1 / t_half)
rad2 = n0 * (0.5) ** (day2 / t_half)

plt.figure(figsize=(10, 6))
plt.plot(time, n_t, label="Radioactive Dose", color="blue")

plt.scatter(day1, rad1, color="green", marker='X', s=100, label="11/9 Radiation (8 days)")
plt.scatter(day2, rad2, color="orange", marker='X', s=100, label="11/17 Radiation (16 days)")

plt.annotate(f"{rad1:.2f} mCi", (day1, rad1), textcoords="offset points", xytext=(50, 0), ha='center', color="green")
plt.annotate(f"{rad2:.2f} mCi", (day2, rad2), textcoords="offset points", xytext=(50, 0), ha='center', color="orange")

plt.title("Radioactive Decay of S-35 Over Time")
plt.xlabel("Time (days)")
plt.ylabel("Dose (mCi)")
plt.grid(True)
plt.legend()
plt.show()


```
